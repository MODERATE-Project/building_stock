# Method 1 - Linear distribution by floor area

```python
import pandas as pd

# Load household floor area

file_path = r"C:\VITO\MODERATE\synthetic_households_all_simulated.xlsx"
df = pd.read_excel(file_path)

# Total annual street-level consumption (in kWh)
total_street_consumption = 3000000  # replace with the actual number from street level consumption, e.g. Fluvius data in BE
```

```python
# Calculate each household's share
df['weight'] = df['floor_area'] / df['floor_area'].sum()

# Multiply by total to get estimated annual household consumption
df['disaggregated_kWh'] = df['weight'] * total_street_consumption

# Save output
df[['household_id', 'disaggregated_kWh']].to_csv("annual_disaggregated.csv", index=False)

print(df[['household_id', 'disaggregated_kWh']])
```

```python
# Save results to Excel
output_path = r"C:\VITO\MODERATE\linear_disaggregation_area.xlsx"
df.to_excel(output_path, index=False)

print("Done! File saved to:", output_path)
```

```python
# Plot histogram
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(df['disaggregated_kWh'], bins=15, edgecolor='black')
plt.title('Histogram of Disaggregated Building Energy Consumption')
plt.xlabel('Disaggregated kWh per Household')
plt.ylabel('Number of Households')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot histogram
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(df['floor_area'], bins=15, edgecolor='black')
plt.title('Histogram of floor_area')
plt.xlabel('floor_area per Household')
plt.ylabel('Number of Households')
plt.grid(True)
plt.tight_layout()
plt.show()
```
